```python
# Get HIV data

import wbpy
api = wbpy.IndicatorAPI()
iso_country_codes = ["KEN", "ZAF", "NGA", "ZWE", "ETH", "UGA",
                     "GHA", "SEN", "MDG", "MOZ", "NAM", "BDI",
                    "BEN", "SOM", "MLI", "AGO", "SLE", "RWA"]
HIV = "SH.DYN.AIDS.ZS"

dataset_HIV = api.get_dataset(HIV, iso_country_codes, date = "2000:2017")
dataset_HIV
```

```python
d = dataset_HIV.as_dict()
```

```python
HIV_df = pd.DataFrame(data = d)
```
