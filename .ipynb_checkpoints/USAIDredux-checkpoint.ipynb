```python
clean1 = usaid.drop(columns= ['country_id', 'country_code', 'region_id',
    'income_group_id', 
       'income_group_acronym', 'implementing_agency_id',
       'implementing_subagency_id', 'subagency_acronym', 
       'channel_category_id', 
       'channel_subcategory_id',  'channel_id',
       'channel_name', 'dac_category_id', 
       'dac_sector_code', 'dac_purpose_code',
      'funding_account_id', 
       'funding_agency_id',  
 'activity_id',
      'activity_project_number', 'activity_start_date',
       'activity_end_date', 'transaction_type_id', 
        'USG_sector_id',
       'submission_id'])
```

```python
clean1.head()
```

```python
clean1.columns
```

```python
clean1 = clean1.drop(columns = ['assistance_category_id','aid_type_group_id'])
```

```python
clean1.columns
```

```python
ssa = clean1.loc[clean1["region_name"]=="Sub-Saharan Africa", :]
```

```python
ssa.head()
```

```python
ssa2 = ssa.loc[(ssa["fiscal_year"]== 2000)|(ssa["fiscal_year"]== 2001)|(ssa["fiscal_year"]== 2002)|
              (ssa["fiscal_year"]== 2003)|(ssa["fiscal_year"]== 2004)|(ssa["fiscal_year"]== 2005)|
              (ssa["fiscal_year"]== 2006)|(ssa["fiscal_year"]== 2007)|(ssa["fiscal_year"]== 2008)|
              (ssa["fiscal_year"]== 2009)|(ssa["fiscal_year"]== 2010)|(ssa["fiscal_year"]== 2011)|
              (ssa["fiscal_year"]== 2012)|(ssa["fiscal_year"]== 2013)|(ssa["fiscal_year"]== 2014)|
              (ssa["fiscal_year"]== 2015)|(ssa["fiscal_year"]== 2016)|(ssa["fiscal_year"]== 2017)]
```

```python
ssa2.head()
```

```python
HIV = ssa2.loc[ssa2["USG_sector_name"]=="HIV/AIDS"]
```
